```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#load dataset
df_customers = pd.read_csv('../dataset/customers_dataset.csv')
df_customers.head()
```

```python
df_orders = pd.read_csv('../dataset/orders_dataset.csv')
df_orders.head()
```

```python
#merge datasets
df = pd.merge(df_customers, df_orders, on='customer_id')
df.head()
```

```python
#create years column from order_purchase_timestamp
df['order_purchase_year'] = pd.to_datetime(df['order_purchase_timestamp']).dt.year
df.head()
```

```python
#lets agregate customer_unique_id and order_purchase_year
df_cust = df.groupby(['order_purchase_year'])['customer_unique_id' ].nunique()
df_cust.head()
```

```python
#lets show the visualization using seaborn
ax = sns.barplot(x=df_cust.index, y=df_cust.values)

ax.set_xlabel('Year')
ax.set_ylabel('Number of customers')
ax.set_title('Number of Customers by Year')

#set annotation
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
```

```python
df['order_purchase_month'] = pd.to_datetime(df['order_purchase_timestamp']).dt.month
df.head()
```

```python
df_actv_month = df.groupby(['order_purchase_month'])['customer_unique_id' ].count()
```

```python
sns.lineplot(x=df_actv_month.index, y=df_actv_month.values)
```
